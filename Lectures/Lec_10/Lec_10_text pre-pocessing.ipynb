## Stemming
- wikipedia: https://en.wikipedia.org/wiki/Stemming
    - stemming is the process of reducing inflected (or sometimes derived) words to their word stem, base or root form—generally a written word form. 
    - The stem need not be identical to the morphological root of the word; 
    - Related words map to the same stem, even if this stem itself is not a valid root
    - **The stem need not to be a word**
    
    
- Examples:
    - cats, catlike --> cat
    - fishing, fished --> fish
    - argue, argued, argues, arguing, and argus --> argu
    - connected, connecting, connection, connections --> connect
    
    
- Application:
    - an approximate method for grouping words with a similar basic meaning together
    - part of a term normalisation process when setting up Information Retrieval systems
    - support search using alternative forms of words
        - Google search adopted a word stemming in 2003
            - Previously a search for "fish" would not have returned "fishing" or "fishes"
    - But in some cases, words with the same morphological stem have different meanings (e.g., cat VS catty)

```python
from nltk.stem import LancasterStemmer
# docs: https://www.nltk.org/api/nltk.stem.html#module-nltk.stem.lancaster

lc_stemmer = LancasterStemmer()
word_list = ['argue', 'argued', 'argues', 'arguing', 'argus'] 
#word_list = ["program", "programs", "programer", "programing", "programers"] 
for wd in word_list:
    print(lc_stemmer.stem(wd))
```

### Porter Stemmer
- one of the oldest, most popular stemming methods proposed in 1980
- remove the common morphological and inflexional endings from words in **English**
- known for its speed and simplicity


- applications:
    - data mining and Information retrieval
    
- limitation:
    - limited to **English** words
    - the stem is not necessarily a meaningful word
  
  
- Resources:
    - official web: https://tartarus.org/martin/PorterStemmer/
    - paper: https://www.cs.odu.edu/~jbollen/IR04/readings/readings5.pdf

```python
# Porter Stemmer
from nltk.stem import PorterStemmer 

pt_stemmer = PorterStemmer() 

word_list = ['connected', 'connecting', 'connection', 'connections']
# word_list = ['argue', 'argued', 'argues', 'arguing', 'argus'] # root: argue; stem: argu
# word_list = ["program", "programs", "programer", "programing", "programers"] 
for wd in word_list:
    print(pt_stemmer.stem(wd)) # get the stem of each word
```

### Snowball Stemmer
- Porter2: also developed by Martin Porter
    - Porter created a programming language called Snowball to processes small strings
    - the most widely used stemmer
    - faster computation speed than PorterStemmer (Porter1)


- multi-lingual stemmer, support the following languages:
    - Arabic, Danish, Dutch, English, Finnish, French, German, Hungarian, Italian, Norwegian, Portuguese, Romanian, Russian, Spanish and Swedish
    - nltk doc: https://www.nltk.org/_modules/nltk/stem/snowball.html#SnowballStemmer


- Resources:
    - http://snowball.tartarus.org/


```python
from nltk.stem import SnowballStemmer
# doc: https://www.nltk.org/api/nltk.stem.html#module-nltk.stem.snowball

print(SnowballStemmer.languages) # See which languages are supported
```

```python
sb_stemmer = SnowballStemmer(language='english') # choose a language
word_list = ['argue', 'argued', 'argues', 'arguing', 'argus'] 
#word_list = ["program", "programs", "programer", "programing", "programers"] 
for wd in word_list:
    print(sb_stemmer.stem(wd))
```

```python
# directly invoke the stemmers for a specific language
from nltk.stem.snowball import EnglishStemmer
en_stemmer = EnglishStemmer()
for wd in word_list:
    print(en_stemmer.stem(wd))
```

```python
from nltk.stem import RegexpStemmer

# uses regular expressions to identify morphological affixes
# any substrings that match the regular expressions will be removed

re_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
word_list = ['argue', 'argued', 'argues', 'arguing', 'argus'] 
# word_list = ["program", "programs", "programer", "programing", "programers"] 
for wd in word_list:
    print(re_stemmer.stem(wd))
```

## Lemmatization
- wikipedia definition:
    - https://en.wikipedia.org/wiki/Lemmatisation
    - the process of grouping together the inflected forms of a word so they can be analysed as a single item, identified by the word's lemma, or dictionary form
    - another process of reducing inflection from words
    - determining the lemma of a word based on its POS tag and intended meaning
    

- Lemmatization VS stemming: 
    - related but different
    - lemmatisation depends on knowledge of the context
        - the intended POS and meaning of a word in a sentence
        - the neighboring sentences or even an entire document
        - reduces words to their origins which have actual meaning 
    - stemmer operates on a single word *without* knowledge of the context
        - cannot discriminate between words that have different meaning depending on part of speech
        - Stemming sometimes generates non-actual words

### Off-the-shelf tagger

```python
from nltk import pos_tag # PerceptronTagger by default
pos_tag(word_tokenize("John's big idea isn't all that bad."))
```

### UnigramTagger

```python
from nltk.corpus import brown
from nltk.tag import UnigramTagger

tagger = UnigramTagger(brown.tagged_sents(categories='news')) # train a unigram tagger
tagger.tag(word_tokenize("John's big idea isn't all that bad."))
```

### HMMTagger

```python
from nltk.tag.hmm import HiddenMarkovModelTagger

hmm_tagger = HiddenMarkovModelTagger.train(brown.tagged_sents(categories='news')) # train a HMM tagger

hmm_tagger.tag(word_tokenize("John's big idea isn't all that bad."))
```

### StanfordPOSTagger

```python
from nltk.tag.stanford import StanfordPOSTagger
# https://nlp.stanford.edu/software/tagger.shtml

stf_tagger = StanfordPOSTagger("english-bidirectional-distsim.tagger","stanford-postagger.jar")

stf_tagger.tag(word_tokenize("John's big idea isn't all that bad."))
```

### BrillTagger: transformation-based learning
    - doc: https://www.nltk.org/_modules/nltk/tag/brill_trainer.html
    - start with an initial tagger
    - uses a series of rules to correct the results of an initial tagger

```python
from nltk.tbl.template import Template
from nltk.tag.brill import Pos, Word
from nltk.tag import untag, RegexpTagger, BrillTaggerTrainer
from nltk.corpus import treebank

# data
training_data = treebank.tagged_sents()[:100]
baseline_data = treebank.tagged_sents()[100:200]
gold_data = treebank.tagged_sents()[200:300]
testing_data = [untag(s) for s in gold_data]


backoff = RegexpTagger([
                        (r'^-?[0-9]+(.[0-9]+)?$', 'CD'),   # cardinal numbers
                        (r'(The|the|A|a|An|an)$', 'AT'),   # articles
                        (r'.*able$', 'JJ'),                # adjectives
                        (r'.*ness$', 'NN'),                # nouns formed from adjectives
                        (r'.*ly$', 'RB'),                  # adverbs
                        (r'.*s$', 'NNS'),                  # plural nouns
                        (r'.*ing$', 'VBG'),                # gerunds
                        (r'.*ed$', 'VBD'),                 # past tense verbs
                        (r'.*', 'NN')                      # nouns (default)
                        ])

baseline = backoff

baseline.evaluate(gold_data) # only use RegexpTagger as backoff 

# templates
Template._cleartemplates() #clear any templates created in earlier tests
templates = [Template(Pos([-1])), Template(Pos([-1]), Word([0]))]

# use BrillTaggerTrainer() to train a BrillTagger with baseline tagger and rules defined in the template
bt_trainer = BrillTaggerTrainer(baseline, templates, trace=3) # RegexpTagger as baseline 
brill_tagger = bt_trainer.train(training_data, max_rules=10)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# generate a simple word cloud from a string.

basecloud = WordCloud().generate(raw_text)

# use matplotlib to show the word cloud
def plot_wordcloud(wordcloud):
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
    
# "your","are", "to", and "all" are missing
# the wordcloud module ignores stopwords by default
plot_wordcloud(basecloud)
```

```python
# the provided stopwords
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
print(stopwords)
```

```python
# specify our own stopwords instead of using the default stopwords
wordcloud = WordCloud(stopwords={'to', 'of'}).generate(raw_text)
plot_wordcloud(wordcloud)
```

```python
# 1789 inaugural address of Washington
washington = nltk.corpus.inaugural.raw('1789-Washington.txt')

# Word cloud for Washington:
wt_wordcloud = WordCloud(relative_scaling=1.0).generate(washington)
plot_wordcloud(wt_wordcloud)
# the most popular words in Washington's inaugural address
```

```python
stopwords.add("every")
stopwords.add("will")
wordcloud = WordCloud(stopwords=stopwords, relative_scaling=1.0).generate(washington)
plot_wordcloud(wordcloud)
```
